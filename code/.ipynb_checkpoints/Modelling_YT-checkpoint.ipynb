```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/train_modified_rows.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
```

```python
df.head(3)
```

```python
train_2007 = df["2007"]
train_2009 = pd.concat([df["2007"], df["2009"]], axis=0)
train_2011 = pd.concat([df["2007"], df["2009"], df["2011"]], axis=0)
train_2013 = df
```

```python
print(train_2007.shape)
print(train_2009.shape)
print(train_2011.shape)
print(train_2013.shape)
```

```python
def drop_targets(df):
    df = df.drop("NumMosquitos", axis=1)
    df = df.drop("WnvPresent", axis=1)
    df = df.drop("wnv_percent", axis=1)
    return df
```

### Predict for 2008:

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

#### Logistic Regression

```python
X_train, X_valid, y_train, y_valid = train_test_split(drop_targets(train_2007), train_2007["WnvPresent"], shuffle=False)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
y_train.value_counts()
```

```python
y_valid.value_counts()
```

```python
dummy_species = pd.get_dummies(X_train["Species"])
dummy_trap = pd.get_dummies(X_train["Trap"])
train_dummies = pd.concat([dummy_species, dummy_trap], axis=1)
train_dummies.shape
```

```python
dummy_species = pd.get_dummies(X_valid["Species"])
dummy_trap = pd.get_dummies(X_valid["Trap"])
valid_dummies = pd.concat([dummy_species, dummy_trap], axis=1)
valid_dummies.shape
```

```python
lr = LogisticRegression(penalty="elasticnet", solver="saga")

params = {
    "l1_ratio": np.linspace(0, 1, 5),
    "C": np.linspace(0.001, 1, 5)
}

gs = GridSearchCV(lr, params, cv=4)
```

```python
%%time
gs.fit(train_dummies, y_train)
```

```python
gs.best_params_
# l1_ratio = 0 means 100% l2 (ridge) penalty
```

```python
print("Training score:", gs.score(train_dummies, y_train))
print("Test score:", gs.score(valid_dummies, y_valid))
```

```python
predictions = gs.predict(valid_dummies)
roc_auc_score(y_valid, predictions) # very poor score of 0.5
```

```python
# model predicts 0 positive WNV, which is a problem
# because of imbalanced classes, model is not sensitive to positive WNV predictions
# we will need to lower the threshold for a positive WNV prediction
```

```python
gs.predict_proba(valid_dummies)
```

```python
prob_wnv_present = pd.DataFrame(gs.predict_proba(valid_dummies))[1]
```

```python
prob_wnv_present.value_counts()
```

```python
plt.boxplot(prob_wnv_present);
```

```python
prob_wnv_present.std()
```

```python
prob_wnv_present.mean()
```

```python
prob_wnv_present.mean() + prob_wnv_present.std() # 1 std above mean
```

```python
def re_predict_std(list_probas, std):
    re_predict = (list_probas > (list_probas.mean() + (std*list_probas.std()))).astype(int)
    return re_predict
```

```python
re_predict_2007 = re_predict_std(list_probas=prob_wnv_present, std=1)
```

```python
re_predict_2007.head()
```

```python
# ROC AUC of cross-validation for data from year 2007
roc_auc_score(y_valid, re_predict_2007) # better score
```

```python
prob_wnv_present.values
```

```python
prob_wnv_present.values.reshape(-1, 1)
```

```python
# for kaggle submission, we need the adjusted probability of present WNV
# adjust probability with standardising
scaler = MinMaxScaler()
# need to reshape because method expects 2D array
prob_wnv_present_scaled = scaler.fit_transform(prob_wnv_present.values.reshape(-1, 1))
prob_wnv_present_scaled
```

```python
[val[0] for val in prob_wnv_present_scaled] # reshape array of arrays to list
```

```python
def scale_1D_probas(list_probas):
    scaler = MinMaxScaler()
    probas_scaled = scaler.fit_transform(list_probas.values.reshape(-1, 1)) # reshape because method expects 2D array
    return [val[0] for val in probas_scaled] # reshape array of arrays to list
```

```python
scale_1D_probas(prob_wnv_present)
```

```python
# custom function just for baseline model
# baseline mode only takes into account species and traps in a time-series
def dummy_species_trap(df):
    dummy_species = pd.get_dummies(df["Species"])
    dummy_trap = pd.get_dummies(df["Trap"])
    df_dummies = pd.concat([dummy_species, dummy_trap], axis=1)
    return df_dummies
```

#### Baseline Model - Compounded Training Sets

```python
# methodology revolves around maximising training data available for use
# in order to prevent data leakage in time-series data:
### data from 2007 is used to predict for 2008
### data from 2007, 2009 is used to predict for 2010
### data from 2007, 2009, 2011 is used to predict for 2012
### data from 2007, 2009, 2011, 2013 is used to predict for 2014
```

```python
def kaggle_baseline_proba(df_train, target, df_test, estimator):
    model = estimator
    df_train_dumm = dummy_species_trap(df_train) # get dummies as baseline predictors
    model.fit(df_train_dumm, df_train[target])
    df_test_dumm = dummy_species_trap(df_test) # get dummies as baseline predictors
    probas = model.predict_proba(df_test_dumm)
    probas = pd.DataFrame(probas)[1] # extract probabilities of positive predictions
    probas = scale_1D_probas(probas) # scale
    df_kaggle = pd.DataFrame(zip(df_test["Id"], probas), columns=["Id", "WnvPresent"])
    return df_kaggle
```

```python
# create dataframes of predictions
df_kaggle_2008 = kaggle_baseline_proba(df_train=train_2007, target="WnvPresent", df_test=test_2008,
                                       estimator=LogisticRegression(penalty="l2", solver="liblinear", C=0.001))
df_kaggle_2010 = kaggle_baseline_proba(df_train=train_2009, target="WnvPresent", df_test=test_2010,
                                       estimator=LogisticRegression(penalty="l2", solver="liblinear", C=0.001))
df_kaggle_2012 = kaggle_baseline_proba(df_train=train_2011, target="WnvPresent", df_test=test_2012,
                                       estimator=LogisticRegression(penalty="l2", solver="liblinear", C=0.001))
df_kaggle_2014 = kaggle_baseline_proba(df_train=train_2013, target="WnvPresent", df_test=test_2014,
                                       estimator=LogisticRegression(penalty="l2", solver="liblinear", C=0.001))

# concat all predictions into one df for Kaggle submission
df_kaggle = pd.concat([df_kaggle_2008, df_kaggle_2010, df_kaggle_2012, df_kaggle_2014]).reset_index(drop=True)
df_kaggle.shape
```
