```python
def replace_m_null(df, col):
    df[col] = df[col].str.replace("M", "99999") # use placeholder numeric value
    df[col] = df[col].astype(float) # convert to float objs
    null_val = df[col] == 99999 # create mask to find placeholders
    df.loc[null_val, col] = np.nan # replace placeholders with true nulls
```

#### Mean Weather

```python
df_weather_mean = pd.DataFrame(columns=df_weather_sel.columns) # create empty dataframe
```

```python
df_weather_mean["Date"] = df_weather_sel.groupby("Station").get_group(1)["Date"] # populate unique observed dates
df_weather_mean.head(3)
```

```python
df_weather_mean.drop("Station", axis=1, inplace=True) # drop "Station" since mean values are station-independent
df_weather_mean.head(3)
```

```python
def mean_weather(df, col):
    mw = [] # create empty list
    for i in np.arange(0, len(df), 2): # create range of indexes in steps of 2
        mw.append(df.loc[i:(i+1), col].mean())
    return mw
```

```python
rem_features = [col for col in df_weather_mean.columns if col != "Date"]
for feat in rem_features:
    df_weather_mean[feat] = mean_weather(df=df_weather_sel, col=feat) # fill all cols with mean values
df_weather_mean.head(3)
```

```python
# case study of trap T115:
df_trap_115 = df_train_date_weather.groupby("Trap").get_group("T115")
df_trap_115.shape
```

```python
# total no. of mosquitos captured by T115 for each date
dict_sum_mozzy = dict(df_trap_115.groupby("Date")["NumMosquitos"].sum())
```

```python
df_trap_115["wnv_product"] = df_trap_115["NumMosquitos"]*df_trap_115["WnvPresent"]
df_trap_115.head()
```

```python
sum_mozzy = []
for date in df_trap_115["Date"]:
    for time, mozzies in dict_sum_mozzy.items():
        if date == time:
            sum_mozzy.append(mozzies)
df_trap_115["sum_mozzy"] = sum_mozzy
df_trap_115.head()
```

```python
# estimate WNV infection rate in mosquito population
df_trap_115["wnv_percent"] = df_trap_115["wnv_product"] / df_trap_115["sum_mozzy"]
df_trap_115.head()
```

```python
# estimated rate of WNV infection in mosquito population
df_trap_115_infection = pd.DataFrame(df_trap_115.groupby("Date")["wnv_percent"].sum())
df_trap_115_infection.head()
```

```python
df_trap_115_infection.reset_index(inplace=True)
df_trap_115_infection.head()
```

```python
df_trap_115_infection["Year"] = df_trap_115_infection["Date"].dt.year # extract year
```

```python
df_trap_115_infection["Month"] = df_trap_115_infection["Date"].dt.month # extract month
```

```python
plt.figure(figsize=[12,8])

plt.title("Rate of WNV infection in Mosquito Population (T115)", size=20)

sns.lineplot(data=df_trap_115_infection, x="Month", y="wnv_percent", hue="Year", ci=False)
plt.ylabel("")

plt.show()
```
